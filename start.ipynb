Algorithmic Solution
> - A* search starting with spots or areas where a set of dominoes must go and trying to chain until we have no more guatanteed sets. A set should correspond to 1 condition of the map.  
> - Within each A* we will perform another search to decide the orientation of the dominoes.  
> - When there is more than 1 valid set of dominoes and/or orientation we pick 1 valid random set/orientation and proceed from there. We mark this step as 1 of ambiguity or potential error.
> - When an error is reached in the map (either a condition is not satisfied or there is no place for a domino to go) we trace back to the last point of ambiguity and restart the search from there ruling out the previous map state.
